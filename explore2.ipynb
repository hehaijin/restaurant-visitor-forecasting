```python
import os.path as path
import pandas as pd
```

```python
input_path="data"
```

```python
air_reserve=pd.read_csv(path.join(input_path,"air_reserve.csv"))
air_visit=pd.read_csv(path.join(input_path,"air_visit_data.csv"))
air_store_info=pd.read_csv(path.join(input_path,"air_store_info.csv"))
date_info= pd.read_csv(path.join(input_path,"date_info.csv"))
store_id_relation=pd.read_csv(path.join(input_path,"store_id_relation.csv"))
hpg_reserve= pd.read_csv(path.join(input_path,"hpg_reserve.csv"))
hpg_store_info=pd.read_csv(path.join(input_path,"hpg_store_info.csv"))
sample_submission=pd.read_csv(path.join(input_path,"sample_submission.csv"))
```

```python
air_visit.head()
```

```python
air_store_info.head()
```

```python
air_visit=pd.merge(air_visit,air_store_info,on=["air_store_id"],how='left')
air_visit.head()
```

```python
air_visit.visit_date=pd.to_datetime(air_visit.visit_date)
```

```python
air_visit["year"]=air_visit.visit_date.dt.year
air_visit["month"]=air_visit.visit_date.dt.month
air_visit["day"]=air_visit.visit_date.dt.day
air_visit.visit_date=air_visit.visit_date.dt.date
air_visit.head()
```

```python
date_info.day_of_week=date_info.day_of_week.astype("category")
date_info.day_of_week=date_info.day_of_week.cat.codes
date_info.calendar_date=pd.to_datetime(date_info.calendar_date)
date_info.calendar_date=date_info.calendar_date.dt.date
```

```python
date_info.head()
```

```python
air_visit=pd.merge(air_visit,date_info,left_on=["visit_date"],right_on=["calendar_date"],how='left').drop("calendar_date",axis=1)
```

```python
air_visit.head()
```

```python
air_visit.air_genre_name=air_visit.air_genre_name.astype("category")
air_visit.air_genre_name=air_visit.air_genre_name.cat.codes
air_visit.air_area_name=air_visit.air_area_name.astype("category")
air_visit.air_area_name=air_visit.air_area_name.cat.codes
air_visit.head()
```

```python
air_visit.dtypes
```

```python
g1=air_visit.groupby(["air_store_id"],as_index=False)["visitors"].mean()

g1.head()
```

```python
g1=g1.rename(columns={"visitors":"average_visitors"})
g1.head()
```

```python
air_visit=pd.merge(air_visit,g1,on=["air_store_id"],how='left')
air_visit.head()
```

```python
sample_submission=pd.merge(sample_submission,date_info,left_on=["visit_date"],right_on=["calendar_date"],how='left')
sample_submission.head()
```
